```python
import pandas as pd
import os
import datetime
import numpy as np
from IPython.display import display
import time
```

```python

def txt_to_csv(txt_file, csv_file, delimiter=';'):

    if os.path.exists(csv_file):
        os.remove(csv_file)  

    df = pd.read_csv(txt_file, delimiter=delimiter)
    df.to_csv(csv_file, index=False)
    df.info()

txt_file_path = 'household_power_consumption.txt'
csv_file_path = 'household_power_consumption.csv'

txt_to_csv(txt_file_path, csv_file_path)

```

```python
def pd_df(csv_file_path):
    dtype = {
        'Global_active_power': float,
        'Global_reactive_power': float,
        'Voltage': float,
        'Global_intensity': float,
        'Sub_metering_1': float,
        'Sub_metering_2': float,
        'Sub_metering_3': float
    }

    df = pd.read_csv(csv_file_path, dtype=dtype, na_values='?')

    df['Date'] = pd.to_datetime(df['Date'], format='%d/%m/%Y')
    df['Time'] = pd.to_datetime(df['Time'], format='%H:%M:%S').dt.time
    df = df.drop(df[(df['Sub_metering_1'] == 0) & (df['Sub_metering_2'] == 0) & (df['Sub_metering_3'] == 0)].index)
    df.dropna(inplace=True)
    return df
csv_file_path = 'household_power_consumption.csv'
df = pd_df(csv_file_path)    
display(df.head())
print(df.isnull().sum())
```

<h3>Перелік атрибутивної інформації:</h3>
<h5>
<ul>
<li>date: дата виміру у форматі dd/mm/yyyy</li>
<li>time: час у форматі hh:mm:ss</li>
<li>global_active_power: активна потужність, яку споживає домогосподарство за хвилину (усереднено) [кВт]</li>
<li>global_reactive_power: реактивна потужність, яку споживає домогосподарство за хвилину (усереднено) [кВт]</li>
<li>voltage: напруга, усереднена за хвилину спостереження [В]</li>
<li>global_intensity: усереднена силу струму для домогосподарства [A]</li>
<li>sub_metering_1: набір споживачів енергії No1 [Вт-годин активної енергії], відповідає кухні, на якій є машина для миття посуду на мікрохвильовка (електричної плити немає, використовується газова).</li>
<li>sub_metering_2: набір споживачів енергії No2 [Вт-годин активної енергії], відповідає пральні, в якій працює пральна машина, сушарка, холодильних та ввімкнено світло.</li>
<li>sub_metering_3: набір споживачів енергії No3 [Вт-годин]</li>
</ul>
</h5>


<h3> 1. Обрати всі домогосподарства, у яких загальна активна споживана потужність перевищує 5 кВт</h3>

```python
def power_over_5(df):
    result_df = df[df['Global_active_power'] > 5.0]
    
    return result_df

start_time = time.time()
filtered_df = power_over_5(df)
result_dataframe = power_over_5(df)
display(result_dataframe)
end_time = time.time() - start_time
print(f"Час виконання Pd: {end_time}")
```

<h3>NumPy:</h3>

```python
def power_over_5_np(df):
    filtered_df = df[np.array(df['Global_active_power'] > 5.0)]
    return filtered_df


start_time = time.time()
result_dataframe = power_over_5_np(df)
display(result_dataframe)
end_time = time.time() - start_time
print(f"Час виконання Numpy: {end_time}")
```

<h3> 2. Обрати всі домогосподарства, у яких вольтаж перевищую 235 В.</h3>

```python
def voltage_over_235(df):

    result_df = df[df['Voltage'] > 235]
    
    return result_df

start_time = time.time()
result_dataframe = voltage_over_235(df)
display(result_dataframe)
end_time = time.time() - start_time
print(f"Час виконання Pd: {end_time}")
```

<h3> 3. Обрати всі домогосподарства, у яких сила струму лежить в межах 19-20 А, для них виявити ті, у яких пральна машина та холодильних споживають більше, ніж бойлер та кондиціонер.</h3>

```python
def within_current_range(df):
    filtered_df = df[(df['Global_intensity'] >= 19) & (df['Global_intensity'] <= 20)]
    
    result_df = filtered_df[(filtered_df['Sub_metering_2'] + filtered_df['Sub_metering_3']) >
                            (filtered_df['Sub_metering_1'] + filtered_df['Sub_metering_3'])]
    
    return result_df

start_time = time.time()
result_dataframe = within_current_range(df)
display(result_dataframe)
end_time = time.time() - start_time
print(f"Час виконання Pd: {end_time}")
```

<h3>NumPy:</h3>

```python
def within_current_range_np(df):
    filtered_df = df[(df['Global_intensity'].values >= 19) & (df['Global_intensity'].values <= 20)]
    
    result_df = filtered_df[(filtered_df['Sub_metering_2'].values + filtered_df['Sub_metering_3'].values) >
                            (filtered_df['Sub_metering_1'].values + filtered_df['Sub_metering_3'].values)]
    
    return result_df

start_time = time.time()
result_dataframe = within_current_range_np(df)
display(result_dataframe.head())
end_time = time.time() - start_time
print(f"Час виконання Pd: {end_time}")
```

<h3>4. Обрати випадковим чином 500000 домогосподарств (без повторів елементів вибірки), для них обчислити середні величини усіх 3-х груп споживання електричної енергії</h3>

```python
def random_house_avarage(df, x=50000):
    random_indices = np.random.choice(df.index, size=x, replace=True)
    random_raw = df.loc[random_indices]
    average_value = random_raw[['Sub_metering_1', 'Sub_metering_2', 'Sub_metering_3']].mean()
    
    return average_value

start_time = time.time()
average_value = random_house_avarage(df)
print(average_value)
end_time = time.time() - start_time
print(f"Час виконання Pd: {end_time}")
```

<h3>NumPy:</h3>

```python
def random_house_avarage_np(df, x=5000):
    random_indices = np.random.choice(df.index, size=x, replace=True)
    random_raw = df.loc[random_indices]
    average_value = np.mean(random_raw[['Sub_metering_1', 'Sub_metering_2', 'Sub_metering_3']], axis=0)
    
    return average_value

start_time = time.time()
average_value = random_house_avarage_np(df)
print(average_value)
end_time = time.time() - start_time
print(f"Час виконання Pd: {end_time}")
```

<h3> 5. Обрати ті домогосподарства, які після 18-00 споживають понад 6 кВт за хвилину в середньому, серед відібраних визначити ті, у яких основне споживання електроенергії у вказаний проміжок часу припадає на пральну машину, сушарку, холодильник та освітлення (група 2 є найбільшою), а потім обрати кожен третій результат із першої половини та кожен четвертий результат із другої половини.</h3>

```python
def select_households(df):
    filtered_df = df[(df['Time'] > datetime.time(18, 0, 0)) & (df['Global_active_power'] > 6.0)]

    final_df = filtered_df[(filtered_df['Sub_metering_2'] > filtered_df['Sub_metering_1']) &
                              (filtered_df['Sub_metering_2'] > filtered_df['Sub_metering_3']) &
                              (filtered_df['Sub_metering_2'] > filtered_df['Sub_metering_2'].mean())]
   
    first_half = final_df.head(len(final_df) // 2)
    second_half = final_df.tail(len(final_df) // 2)
    result_df = pd.concat([first_half.iloc[::3, :], second_half.iloc[::4, :]])
    return result_df

start_time = time.time()
result = select_households(df)
display(result)
end_time = time.time() - start_time
print(f"Час виконання Pd: {end_time}")
```

<h3>NumPy:</h3>

```python
def select_households_np(df):
    filtered_df = df[np.array(df['Time'] > datetime.time(18, 0, 0)) & np.array(df['Global_active_power'] > 6.0)]

    final_df = filtered_df[np.array(filtered_df['Sub_metering_2'] > filtered_df['Sub_metering_1']) &
                              (filtered_df['Sub_metering_2'] > filtered_df['Sub_metering_3']) &
                              (filtered_df['Sub_metering_2'] > filtered_df['Sub_metering_2'].mean())]
   
    first_half = final_df.head(len(final_df) // 2)
    second_half = final_df.tail(len(final_df) // 2)
    result_df = pd.concat([first_half.iloc[::3, :], second_half.iloc[::4, :]])
    return result_df

start_time = time.time()
result = select_households_np(df)
display(result)
end_time = time.time() - start_time
print(f"Час виконання Pd: {end_time}")
```
